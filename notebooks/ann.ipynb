```python
import pandas as pd
def decouple(s):
    ls = list(s.split("', '"))
    if len(ls) == 3:
        ls[0] = ls[0][2:]
        ls[2] = ls[2][:-2]
    elif len(ls) == 2:
        ls[0] = ls[0][2:]
        ls[1] = ls[1][:-2]
    dc = {s.split(': ')[0]: s.split(': ')[1] for s in ls}
    return dc

new_ann = pd.DataFrame(columns=['some_id', 'sample_title', 'tissue', 'time point', 'treatment response'])
for i in range(len(ann)):
    new_row = decouple(ann.iloc[i]['characteristics_ch1'])
    title = ann.iloc[i]['title']
    new_row['some_id'] = title[title.find('[', 1) + 1 : title.find(']')]
    new_row['sample_title'] = title[2:title.find(' ')]
    new_ann = pd.concat([new_ann, pd.DataFrame([new_row], index=[ann.index[i]])])

new_ann
```

```python
new_ann.to_csv('../data/processed_ann.csv')
```
