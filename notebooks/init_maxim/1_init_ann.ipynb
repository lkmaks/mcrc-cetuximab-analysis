```python
import pandas as pd
def decouple(s):
    ls = list(s.split("', '"))
    if len(ls) == 3:
        ls[0] = ls[0][2:]
        ls[2] = ls[2][:-2]
    elif len(ls) == 2:
        ls[0] = ls[0][2:]
        ls[1] = ls[1][:-2]
    dc = {s.split(': ')[0]: s.split(': ')[1] for s in ls}
    return dc

def pad_zeros(s, n):
    while len(s) < n:
        s = '0' + s
    return s

new_ann = pd.DataFrame(columns=['some_id', 'sample_title', 'tissue', 'time point', 'treatment response'])
for i in range(len(ann)):
    new_row = decouple(ann.iloc[i]['characteristics_ch1'])
    title = ann.iloc[i]['title']
    new_row['some_id'] = title[title.find('[', 1) + 1 : title.find(']')]
    new_row['sample_title'] = title[2:title.find(' ')]
    # new_row['sample_prefix'] = (lambda s: s[:s.find('-')])(new_row['sample_title'])
    # new_row['patient_id'] = (lambda s: pad_zeros(''.join(filter(lambda x: x.isdigit(), s)), 4))(new_row['sample_title'])
    new_ann = pd.concat([new_ann, pd.DataFrame([new_row], index=[ann.index[i]])])

ann = new_ann
```

```python
def extend_id(id):
    fp, sp = id.split('_')
    while len(sp) < 4:
        sp = '0' + sp
    return fp + '_' + sp

ann['gsm_id'] = ann.index
ann = ann.reset_index(drop=True)
ann = ann.rename(columns={'some_id': 'sample_id'})
ann['sample_id'] = ann['sample_id'].apply(extend_id)
ann
```
